```python
import json
import uuid
import os
import json
from dotenv import load_dotenv
from pathlib import Path
from kafka import KafkaConsumer
from faker import Faker
from time import sleep
from kafka import TopicPartition
```

```python
dotenv_path = Path('/resources/.env')
load_dotenv(dotenv_path=dotenv_path)
```

```python
kafka_host = os.getenv('KAFKA_HOST')
kafka_topic_partition = "nasabah_bank"
```

```python
consumer = KafkaConsumer(
    kafka_topic_partition,
    group_id='nasabah-consumer-group',
    bootstrap_servers=[f'{kafka_host}:9092'],
    value_deserializer=lambda m: json.loads(m.decode('utf-8')),
    fetch_min_bytes=50000,  # Minimum bytes yang harus diambil
    fetch_max_wait_ms=500,  # Maksimal waktu tunggu dalam ms
    auto_offset_reset='earliest',
    enable_auto_commit=False
)
```

```python
# Mengkonsumsi dan memproses data
total_saldo = 0
total_nasabah = 0

commit_batch_size = 10

for message in consumer:
    data = message.value
    print(f"Received message: {data}")
    
    # Perhitungan: Total saldo dari nasabah
    saldo = data['saldo']
    total_saldo += saldo
    total_nasabah += 1
    
    # Menghitung rata-rata saldo
    avg_saldo = total_saldo / total_nasabah if total_nasabah > 0 else 0
    print(f"Rata-rata Saldo Nasabah: {avg_saldo}")

    # Commit offset setelah memproses
    consumer.commit()
```
